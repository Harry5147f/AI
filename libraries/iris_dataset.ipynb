```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
'''Importing iris model in df'''

df=sns.load_dataset("iris")
#Display basic information
print(df.info())
#info = column_names +not null count+ datatype

```

```python
#List of sepal length only of all 150(50+50+50) species
print(df["sepal_length"].value_counts())
```

```python
#Display info of all column
print(df["sepal_length"].mean())
print(f"{df.sepal_length.mean():.2f}")
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
#Total unique names(Though they are repeated but only these 3 are repeated not other)
print(df["species"].value_counts())

```

```python

plt.figure(figsize=(8,5))
sns.countplot(y="species",data=df)
plt.show()
sns.pairplot(df, hue="species")
corr=df.corr(numeric_only=True)

plt.title("Hello")
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
"""
corr le relation dekauxa
annot le information dinxa
cmap le color dinxa. Where cmap=colour map
"""
```
